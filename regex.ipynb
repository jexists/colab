### Search method

→ 첫번째 패턴을 찾으면 match 객체 반환
→ 패턴을 찾지 못하면 None 반환

```python
import re
```

```python
re.search(r'abc', 'abcdef')
```

```python
m = re.search(r'abc', 'abcdef')

print(type(m))
print(m.start()) # 시작 인덱스
print(m.end()) # 끝 인덱스
print(m.group())
```

```python
m = re.search(r'abc', 'abdef')

print(type(m))
print(m)

# 없는 경우 에러 
print(m.start()) # 시작 인덱스
print(m.end()) # 끝 인덱스
print(m.group())
```

```python
# 연달아 나오는 숫자
m= re.search(r'\d\d','112abcdef119')
print(m)
```

```python
m= re.search(r'\d\d\d\w','112abcdef119')
print(m)
```

```python
m= re.search(r'..\w\w','@#$%#$ABCDabcd')
print(m)
```
