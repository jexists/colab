### Numpy

→ Numerical Python

→ 산술계산을 위한 패키지

→ Numpy의 배열 객체를 데이터 교환을 위한 공통 언어처럼 사용

→ ndarray라고 하는 N차원의 배열 객체 (대규모 데이터 집합을 담을 수 있는 자료구조)

```python
import numpy as np
```

```python
a = np.array([1,2,3])
print(a, type(a))
```

```python
# 뒤에 .을 넣을 경우 다 float으로 변환
b = np.array([1,2,3.])
print(b, type(b))
```

```python
# 리스트 vs Numpy

l = [1,2,3]
a = np.array([1,2,3])
print(l)
print(a) # numpy: print로 출력시 컴마x
a
```

```python
for item in l:
  print(item)
```

```python
for item in a:
  print(item)
```

```python
l.append(4)
print(l)
```

```python
# numpy append 지원 x
a.append(4)
print(a)
```

```python
a = np.append(a, 4)
print(a)
```

```python
print(l)
l = l + [5]
print(l)
```

```python
# 브로드케스팅
print(a)
a = a + np.array([5])
print(a)
```

```python
# list *2
l = [1,2,3]
l2 = []
for item in l: 
  l2.append(item*2)
print(l2)
```

```python
# ndarray *2
l = [1,2,3]
a = np.array(l)
a2 = a*2
print(a2)
```

```python
l = [1,2,3]
print(l)
l = l * 2
print(l)
```
