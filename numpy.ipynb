→ Numerical Python

→ 산술계산을 위한 패키지

→ ndarray라고 하는 N차원의 배열 객체 (대규모 데이터 집합을 담을 수 있는 자료구조)

```python
import numpy as np
```

```python
a = np.array([1,2,3])
print(a, type(a))
```

```python
# 리스트 vs Numpy

l = [1,2,3]
a = np.array([1,2,3])
print(l)
print(a) # numpy: print로 출력시 컴마x
a
```

```python
for item in l:
  print(item)
```

```python
for item in a:
  print(item)
```

```python
l.append(4)
print(l)
```

```python
# numpy append 지원 x
a.append(4)
print(a)
```

```python
a = np.append(a, 4)
print(a)
```

```python
print(l)
l = l + [5]
print(l)
```

```python
# 브로드케스팅
print(a)
a = a + np.array([5])
print(a)
```

```python
# list *2
l = [1,2,3]
l2 = []
for item in l: 
  l2.append(item*2)
print(l2)
```

```python
# ndarray *2
l = [1,2,3]
a = np.array(l)
a2 = a*2
print(a2)
```

```python
l = [1,2,3]
print(l)
l = l * 2
print(l)
```

# numpy 배열

→ 모두 같은 유형의 값

→ 정수의 튜플로 인덱싱

→ 차원의 수는 배열의 랭크

→ 배열의 shape은 각 차원별 배열 크기의 튜플

→ 중첩된 파이썬 리스트로부터 배열 초기화

→ 대괄호 사용하여 요소에 접근 가능

```python
a = np.array([1,2,3])
print(a)
print(type(a), a.ndim, a.shape, a.dtype)
```

asarray vs array

```python
# 같은 어레이
b = np.asarray(a)
print(b)
```

```python
b[0] = 9
print(a)
print(b)
```

```python
# 새로운 카피
a = np.array([1,2,3])
b = np.array(a)
print(b)
```

```python
b[0] = 9
print(a)
print(b)
```

### 배열 생성

```python
a = np.zeros((3,4))
print(a)
print(a.dtype)
```

```python
b = np.ones((5,2))
print(b)
print(b.dtype)
```

```python
c = np.full((2,3),4)
print(c)
```

```python
d = np.empty((3,3))
print(d)
```

```python
a = np.array([1,2,3])
a = np.zeros((3,4))
e = np.zeros_like(a)
f = np.ones_like(a)
g = np.full_like(a, 2)
h = np.empty_like(a)
print(a)
print("-------")
print(e)
print(f)
print(g)
print(h)
```

```python
i = np.eye(3)
j = np.identity(3)
print(i)
print(j)
```

```python
k = np.arange(10)
print(k)
k = np.arange(1, 10)
print(k)
k = np.arange(10, 0, -1)
print(k)
```

```python
# uniform distribution
l = np.random.rand(2, 2)
print(l)
```

```python
# normal distribution
m = np.random.randn(2,2)
print(m)
```
