### 배열 (Array)

- 데이터 나열하고, 각 데이터를 인덱스에 대응하도록 구성한 데이터 구조

- 파이썬에서는 리스트 타입이 배열 기능을 제공하고 있음

- 같은 종류의 데이터를 효율적으로 관리하기 위해 사용

- 같은 종류의 데이터를 순차적으로 저장

- 장점: 빠른 접근 가능

- 단점: 추가, 삭제가 어려움

- 미리 최대 길이를 지정 필요

```python
# [C언어]
#include <stdio.h>

int main(int argc, char *argv[]) {
    char country[3] = 'US';
    printf("%c%c\n", country[0], country[1]);
    printf("%s\n",country);
    return 0
}
```

```python
country = "US"
print(country)
country = country + "A"
print(country)
```

```python
# 1차원 배열: 리스트 구현

data = [1,2,3,4,5]
print(data)
```

```python
# 2차원 배열: 리스트 구현

data = [[1,2,3],[4,5,6],[7,8,9]]
print(data)
print(data[0])
print(data[0][0])
```
